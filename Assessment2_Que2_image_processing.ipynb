```python
# Import the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Conv2D
import skimage.measure 
```

```python
#Importing the image

from PIL import Image
image=Image.open('C:\\Users\\Sheetal\\Desktop\\Old Stuff\\nature.jpg')
plt.imshow(image)
```

```python
#resizing image
rsize = image.resize((np.array((15,15))).astype(int))
image = np.asarray(rsize)
image.shape
```

```python
#Puting the image into an array
x=image.reshape(1,15,15,3)
```

```python
def Convolution(x,kernel_size,strides=1, padding='valid'):
 model=Sequential()
 model.add(Conv2D(1, kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(15,15,3)))
 yhat = model.predict(x)
 y=yhat.reshape(yhat.shape[1],yhat.shape[2])
 filtered_image = Image.fromarray(y)
 return filtered_image
```

```python
#As per as the question given above
Filtered_image_11=Convolution(x,(3,3))
Filtered_image_12=Convolution(x,(5,5))
Filtered_image_21=Convolution(x,(3,3),2)
Filtered_image_22=Convolution(x,(5,5),2)
Filtered_image_31=Convolution(x,(3,3),1,'same')
Filtered_image_32=Convolution(x,(5,5),1,'same')
```

## 1.Filter size (3*3)

```python
plt.imshow(Filtered_image_11)
entropy = skimage.measure.shannon_entropy(Filtered_image_11)
print("Entropy:",entropy)
```

## 2.Filter size (5*5)

```python
plt.imshow(Filtered_image_12)
entropy = skimage.measure.shannon_entropy(Filtered_image_12)
print("Entropy:",entropy)
```

## 3.Filter size (3*3), stride=2

```python
plt.imshow(Filtered_image_21)
entropy = skimage.measure.shannon_entropy(Filtered_image_21)
print("Entropy:",entropy)
```

## 4.Filter size (5*5), stride=2

```python
plt.imshow(Filtered_image_22)
entropy = skimage.measure.shannon_entropy(Filtered_image_22)
print("Entropy:",entropy)
```

```python
plt.imshow(Filtered_image_31)
entropy = skimage.measure.shannon_entropy(Filtered_image_31)
print("Entropy:",entropy)
```
